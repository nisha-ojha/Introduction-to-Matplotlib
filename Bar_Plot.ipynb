```python
import matplotlib.pyplot as plt

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.bar(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
```

```python
import numpy as np
import matplotlib.pyplot as plt


sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

years =np.arange(2011, 2021)

x = np.arange(len(years))  # index positions
width = 0.25

plt.bar(x - width, sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width, kohli, width=width, label="Kohli")

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt

players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]
```

```python
#plotting with barh() for Horizontal Bar Charts

plt.barh(players, runs_5yrs, color = "purple")
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
 
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]
plt.bar(players, runs, color="skyblue")

# Add labels on top of bars
for i in range(len(players)):
    plt.text(i, runs[i] + 30, str(runs[i]), ha = 'center')

plt.ylabel('Runs')
plt.title('Runs Scored by Players')
plt.tight_layout()
```

```python
#Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.
#Then create a horizontal bar chart showing total runs scored in their debut 5 years.

years = [2001,2003,2005,2007,2009]
sachin = [800, 900, 700, 1100, 950]
kohli  = [0,   0,   500, 900, 1200]
sehwag = [300, 850, 1100, 1400, 1000]

total_runs = [4550, 3700, 4200]   # sachin, kohli, sehwag
players = ["Sachin", "Kohli", "Sehwag"]

x = np.arange(len(years))
width = 0.25

plt.bar(x - width, sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width, kohli, width=width, label="Kohli")
plt.xticks(x, years)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs Comparison in 5 Selected Years")
plt.legend()
plt.tight_layout()
```

```python
plt.barh(players, total_runs)
plt.xlabel("Total Runs in First 5 Years")
plt.title("Debut 5-Year Performance")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [15, 25, 25, 30, 5]
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']

# Customizing all wedges with a black border
plt.pie(data, labels=labels, wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'fill' : False})
plt.title('Pie Chart with Wedge Borders')
plt.show()

# Creating a donut chart
plt.pie(data, labels=labels, wedgeprops={'width': 0.4, 'edgecolor': 'white'})
plt.title('Donut Chart')
```

```python
runs = [10, 20, 30, 40]
labels = ['Sachin', 'Kohli', 'Sehwag', 'Rohit']
colors = ['red', 'blue', 'green', 'orange']
explode = [0.1, 0, 0, 0]

plt.pie(runs, labels=labels, explode=explode, colors=colors, shadow = True)
plt.title("Exploded Pie Example")
plt.show()
```
