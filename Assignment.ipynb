```python
import matplotlib.pyplot as plt
```

```python
# Hypothetical runs
years = list(range(2011, 2021))

kohli_runs = [900, 1100, 1200, 1000, 1300, 1400, 1550, 1600, 1500, 1700]
rohit_runs = [700, 850, 920, 960, 1100, 1250, 1300, 1450, 1480, 1550]
sehwag_runs = [1050, 1020, 980, 950, 700, 600, 500, 450, 300, 250]

plt.style.use("seaborn-v0_8-darkgrid")  # makes plot clean and readable
plt.figure(figsize=(10,5))  # width x height

plt.plot(years, kohli_runs, color="red", linestyle="-", marker="o", label="Virat Kohli")
plt.plot(years, rohit_runs, color="blue", linestyle="--", marker="^", label="Rohit Sharma")
plt.plot(years, sehwag_runs, color="green", linestyle=":", marker="s", label="Virender Sehwag")

plt.grid(True)
plt.tight_layout()

plt.title("Hypothetical Runs Over 10 Seasons (20102019)")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()
```
